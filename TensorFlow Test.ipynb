```python
import pandas as pd
from sklearn.model_selection import train_test_split

df = pd.read_csv('Churn.csv')
```

```python
x = pd.get_dummies(df.drop(['Churn', 'Customer ID'], axis = 1))
y = df['Churn'].apply(lambda x: 1 if x=='Yes' else 0)

```

```python
x_train, x_test, y_train, y_test = train_test_split(x, y,test_size=.2)
```

```python
from tensorflow.keras.models import Sequential, load_model
from tensorflow.keras.layers import Dense
from sklearn.metrics import accuracy_score
```

```python
model = Sequential()
model.add(Dense(units=32, activation='relu', input_dim=len(x_train.columns)))
model.add(Dense(units=64, activation='relu'))
model.add(Dense(units=1, activation='sigmoid'))
```

```python
#compiling our model using three key values, loss, optimizer and metrics. Loss is the sum of how far our estimations are to sink a battleship. 
# Optimizer is how we choose to search for the battleship i.e. in a grid pattern or zigzag
#metrics is how we evaluate how our model is performing
model.compile(loss='binary_crossentropy', optimizer='sgd', metrics='accuracy')
```

```python
model.fit(x_train, y_train, epochs=200, batch_size=32)
```
