```python
import pandas as pd
from sklearn.model_selection import train_test_split

df = pd.read_csv('Churn.csv')
```

```python
x = pd.get_dummies(df.drop(['Churn', 'Customer ID'], axis = 1))
y = df['Churn'].apply(lambda x: 1 if x=='Yes' else 0)

```

```python
x_train, x_test, y_train, y_test = train_test_split(x, y,test_size=.2)

print(x_test)
```

```python
x_train.head()
```

```python
y_train.head()
```

```python
from tensorflow.keras.models import Sequential, load_model
from tensorflow.keras.layers import Dense
from sklearn.metrics import accuracy_score
```

```python
model = Sequential()
model.add(Dense(units=32, activation='relu', input_dim=len(x_train.columns)))
model.add(Dense(units=64, activation='relu'))
model.add(Dense(units=1, activation='sigmoid'))
```

```python
#compiling our model using three key values, loss, optimizer and metrics. Loss is the sum of how far our estimations are to sink a battleship. 
# Optimizer is how we choose to search for the battleship i.e. in a grid pattern or zigzag
#metrics is how we evaluate how our model is performing
model.compile(loss='binary_crossentropy', optimizer='sgd', metrics='accuracy')
```

```python
#Issues with running the model because some data types in the x_train/y_train were not integer or floats which didn't allow the model to run. We had to convert these into integers/floats by
#converting to numpy

Import numpy as np

print(f'x_train type: {type(x_train)}, shape: {x_train.shape}')
print(x_train.dtypes)
print(f'y_train type: {type(y_train)}, shape: {y_train.shape}')
print(y_train.dtypes)

# Convert to NumPy arrays with appropriate dtype
x_train = x_train.to_numpy(dtype=np.float32)
y_train = y_train.to_numpy(dtype=np.float32)

# Print to verify
print(f'x_train type: {type(x_train)}, dtype: {x_train.dtype}, shape: {x_train.shape}')
print(f'y_train type: {type(y_train)}, dtype: {y_train.dtype}, shape: {y_train.shape}')


# Inspect the NumPy arrays after conversion
print(f'After conversion - x_train type: {type(x_train)}, dtype: {x_train.dtype}, shape: {x_train.shape}')
print(f'After conversion - y_train type: {type(y_train)}, dtype: {y_train.dtype}, shape: {y_train.shape}')

```

```python
model.fit(x_train, y_train, epochs=300, batch_size=32)
#Higher the Epoch the more accurate the model but may lead to overfitting.
#As model trains, the accuracy should be improcing and loss decreasing.
```

```python
y_hat = model.predict(x_test)
y_hat = [0 if val <0.5 else 1 for val in y_hat]
y_hat
```
